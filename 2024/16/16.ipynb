```python
# Separate by line and spaces and turn into integer
import re
s = []

with open('./input.txt', 'r') as file:
    file_content = file.read().strip()  
individual_strings = file_content.split('\n')

for i, individual_str in enumerate(individual_strings, 1):
    word = [*individual_str]
    s.append(word)
```

```python
for i in range(len(s)):
    for j in range(len(s[i])):
        if s[i][j]=='S':
            start = (i,j)
        if s[i][j]=='E':
            end = (i,j)
            
print(start,end)
```

```python
def dir(way,i,j,d):
    if way== d:
        new_c = 0
    if way!=d:
        new_c = 1
        
    if way =='U':
        return((way, i-1,j,new_c)) 
    if way =='D':
        return((way, i+1,j,new_c))
    if way =='L':
        return((way, i,j-1,new_c))
    if way =='R':
        return((way, i,j+1,new_c))
```

```python
def shortest_path(s, start, end):
    # Data structure: (cost, dir, x, y)
    start = (0, 'R', start[0], start[1])  # Starting position

    ol = [start]  # Open List
    cl = set()  # Closed List
    
    costs = {}
    costs[('R', start[0], start[1])] = 0  # Store as (direction, x, y)

    # While there are elements in the open list
    while ol:
        ol.sort()  # Sort the open list based on cost
    
        # We remove it from the open list and add it to the closed list
        cost, d, i, j = ol.pop(0)  
        active = (d, i, j)
        cl.add(active)
        
        # If we reached the end
        if (i, j) == end:
            return cost, active

        directions = ['U', 'D', 'L', 'R']

        if d == 'U':
            directions.remove('D')
        elif d == 'D':
            directions.remove('U')
        elif d == 'R':
            directions.remove('L')
        elif d == 'L':
            directions.remove('R')

        for way in directions:
            if s[dir(way,i,j,d)[1]][dir(way,i,j,d)[2]] != '#':
                new_state = dir(way, i, j, d)
                new_pos = (new_state[1], new_state[2])
            
                if 0 <= new_pos[0] < len(s) and 0 <= new_pos[1] < len(s[0]):
                    new_cost = cost + 1 + (1000 if way != d else 0)
                    new_key = (new_state[0], new_state[1], new_state[2])
                
                    if new_key not in costs or new_cost < costs[new_key]:
                        costs[new_key] = new_cost
                        ol.append((new_cost, new_state[0], new_state[1], new_state[2]))
    
    return None 
```

```python
sp = shortest_path(s, start,end)[0]
sp
```

```python
def bv(s, start, end,sp,best_views):
    
    # Data structure: (cost, dir, x, y)
    start = (0, 'R', start[0], start[1])  # Starting position

    ol = [start]  # Open List
    cl = set()  # Closed List
    
    costs = {}
    costs[('R', start[0], start[1])] = 0  # Store as (direction, x, y)

    path = {}
    path[('R',start[2], start[3])] = {(start[2], start[3])}
    
    # While there are elements in the open list
    while ol:
        ol.sort()  # Sort the open list based on cost
    
        # We remove it from the open list and add it to the closed list
        cost, d, i, j = ol.pop(0)  
        active = (d, i, j)
        cl.add(active)
        
        
        # If we reached the end
        if (i, j) == end and cost == sp:
            for e in path[('U', end[0], end[1])]:
                best_views.add(e)
            
        directions = ['U', 'D', 'L', 'R']

        if d == 'U':
            directions.remove('D')
        elif d == 'D':
            directions.remove('U')
        elif d == 'R':
            directions.remove('L')
        elif d == 'L':
            directions.remove('R')

        for way in directions:
            if s[dir(way,i,j,d)[1]][dir(way,i,j,d)[2]] != '#':
                new_state = dir(way, i, j, d)
                new_pos = (new_state[1], new_state[2])
            
                if 0 <= new_pos[0] < len(s) and 0 <= new_pos[1] < len(s[0]):
                    new_cost = cost + 1 + (1000 if way != d else 0)
                    new_key = (new_state[0], new_state[1], new_state[2])
                    new_key_path = (new_state[0], new_state[1], new_state[2])
                    
                        
                    if new_key not in costs or new_cost <= costs[new_key]:
                        if new_key_path not in path:
                            path[new_key_path] = set(path[(d,i,j)])
                    
                        else:
                            for e in path[(d,i,j)]:
                                path[new_key_path].add(e)
                    
                        path[new_key_path].add(new_pos)
                        costs[new_key] = new_cost
                        ol.append((new_cost, new_state[0], new_state[1], new_state[2]))
    
    return best_views 
```

```python
best_views = set()
res = len(bv(s, start, end, sp, best_views))
print(res)
```
