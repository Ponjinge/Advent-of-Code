```python
import re

lin = []

with open('./input.txt', 'r') as file:
    file_content = file.read().strip()  # Read the file content and remove leading/trailing whitespaces

# Split the content based on empty lines
individual_strings = file_content.split('\n')

# Process each individual
for i, individual_str in enumerate(individual_strings, 1):
    # Convert lines to integers and add them to the individuals list
    individual = [line for line in re.split(" ", individual_str)]
    
    
    lin.append(individual)


    
```

```python
for i in range (0, len(lin)):
    lin[i][0] = lin[i][0].replace('A','Z')
    lin[i][0]=lin[i][0].replace('K', "Y")
    lin[i][0]=lin[i][0].replace('Q', "X")
    lin[i][0]=lin[i][0].replace('J', "0")
    lin[i][0]=lin[i][0].replace('T', "G")
    
```

```python
t5=[]
t4=[]
tfh=[]
t3=[]
td2=[]
t2=[]
thc=[]
for i in range (len(lin)):
    stri = lin[i][0]
    a=[]
    b=[]
    c=[]
    d=[]
    
    jc=0
    for j in range (0,5):
        if stri[j]=='0':
            jc=jc+1
        elif len(a)<1 or stri[j]==a[0]:
            a.append(stri[j])
        elif len(b)<1 or stri[j]==b[0]:
            b.append(stri[j])
        elif len(c)<1 or stri[j]==c[0]:
            c.append(stri[j])
        elif len(d)<1 or stri[j]==d[0]:
            d.append(stri[j])
    
    if (len(a)+ jc==5) or (len(b)+jc==5) or (len(c)+jc==5) or (len(d)+jc==5):
        t5.append(lin[i])
    elif (len(a)+ jc==4) or (len(b)+jc==4) or (len(c)+jc==4) or (len(d)+jc==4):
        t4.append(lin[i])
    elif (len(a)==3 and len(b)==2) or (len(a)==2 and len(b)==3):
        tfh.append(lin[i])
    elif (len(a)+jc+len(b)==5) or (len(a)+jc+len(c)==5) or (len(c)+jc+len(b)==5) or (len(c)+jc+len(d)==5) or (len(a)+jc+len(d)==5) or (len(b)+jc+len(d)==5) :
        tfh.append(lin[i])
    elif (len(a)+ jc==3) or (len(b)+jc==3) or (len(c)+jc==3) or (len(d)+jc==3):
        t3.append(lin[i])
    elif (len(a)==2 and len(b)==2) or (len(a)==2 and len(c)==2) or (len(c)==2 and len(b)==2):
        td2.append(lin[i])
    elif (len(a)+jc+len(b)==4) or (len(a)+jc+len(c)==4) or (len(c)+jc+len(b)==4) or (len(c)+jc+len(d)==4) or (len(a)+jc+len(d)==4) or (len(b)+jc+len(d)==4) :
        td2.append(lin[i])
    elif (len(a)+ jc==2) or (len(b)+jc==2) or (len(c)+jc==2) or (len(d)+jc==2):
        t2.append(lin[i])
        
    else:
        thc.append(lin[i])
        
    
```

```python
t5.sort()
t4.sort()
tfh.sort()
t3.sort()
td2.sort()
t2.sort()
thc.sort()
```
