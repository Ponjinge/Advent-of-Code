```python
#Boilerplate 1 

s = []

with open('./input.txt', 'r') as file:
    file_content = file.read().strip()  
individual_strings = file_content.split('\n')
for i, individual_str in enumerate(individual_strings, 1):
    word = [*individual_str]
    s.append(word)
```

```python
for i in range(len(s)):
    for j in range(len(s[0])):
        s[i][j]= int(s[i][j])
```

```python
def dir(way,i,j,c,d):
    if way== d:
        new_c = c+1
    if way!=d:
        new_c = 1
        
    if way =='U':
        return((way, i-1,j,new_c)) 
    if way =='D':
        return((way, i+1,j,new_c))
    if way =='L':
        return((way, i,j-1,new_c))
    if way =='R':
        return((way, i,j+1,new_c))
```

```python

def shortest_path(s, target):

    # Data structure: (dir, x,y, count)
    start = ('U',0,0,1) # Starting position 

    ol = [start] # Open List
    cl = [] # Closed List 


    # Create an empty dict for the costs to get to a square
    costs = {}
    costs[start] = 0

    # While there are elements in the open list 
    while ol:
    
        #print(ol)
        
        min = 10**10 
        for i in range (len(ol)):
        
            # We treat the lowest cost element in the list
            if costs[ol[i]] < min:
                min = costs[ol[i]]
                min_index = i
            
        active = ol[min_index] # It becomes the active space 
    
        # We remove it from the open list and add it to the closed list
        cl.append(active)
        ol.pop(min_index) 
    
        d, i, j, c = active
        
        if i == target[0] and j == target[1]:
            return costs[active]
        
        # For each possible direction around that space
    
        directions = ['U','D','L','R']
    
        if d == 'U':
            directions.remove('D')
        
        if d == 'D':
            directions.remove('U')
            
        if d == 'R':
            directions.remove('L')
            
        if d == 'L':
            directions.remove('R')
            
        if c >= 3:
            directions.remove(d)
   
        for way in  directions :
  
            #If we reach the end of the grid
            if dir(way,i,j,c,d)[1] == target[0] and dir(way,i,j,c,d)[2] == target[1]:
                return(s[dir(way,i,j,c,d)[1]][dir(way,i,j,c,d)[2]] + costs[active], active)
            
        
    
            # If this square is already in the Open List for this direction
            if dir(way,i,j,c,d) in ol:
            
                if costs[dir(way,i,j,c,d)] >= s[dir(way,i,j,c,d)[1]][dir(way,i,j,c,d)[2]] + costs[active]:
                    
                    costs[dir(way,i,j,c,d)]  = s[dir(way,i,j,c,d)[1]][dir(way,i,j,c,d)[2]] + costs[active]
                
                    
                    
            # If this square is already in the Closed List for this direction            
            elif dir(way,i,j,c,d) in cl:
             
                if costs[dir(way,i,j,c,d)] >= s[dir(way,i,j,c,d)[1]][dir(way,i,j,c,d)[2]] + costs[active]:
                    
                    costs[dir(way,i,j,c,d)]  = s[dir(way,i,j,c,d)[1]][dir(way,i,j,c,d)[2]] + costs[active]
              
                    # Add to open list
                    ol.append(dir(way,i,j,c,d))
               
                    
                    
            # If this space has not been observed yet from this direction           
            elif 0 <= dir(way,i,j,c,d)[1] < len(s) and 0 <= dir(way,i,j,c,d)[2] < len(s[0]):
           
                costs[dir(way,i,j,c,d)]  = s[dir(way,i,j,c,d)[1]][dir(way,i,j,c,d)[2]] + costs[active]
            
                # Add to open list
                ol.append(dir(way,i,j,c,d))
            
       
                
```

```python
def shortest_path(s, target):
    # Data structure: (cost, dir, x, y, c)
    start = (0, 'U', 0, 0, 0)  # Starting position

    ol = [start]  # Open List
    cl = set()  # Closed List

    
    costs = {}
    costs[start[1:]] = 0

    # While there are elements in the open list
    while ol:
        ol.sort()  # Sort the open list based on cost
    
        # We remove it from the open list and add it to the closed list
        cost, d, i, j, c = ol.pop(0)  
        active = (cost, d, i, j, c)
        cl.add(active[1:])
        
        directions = ['U', 'D', 'L', 'R']

        if d == 'U':
            directions.remove('D')
        elif d == 'D':
            directions.remove('U')
        elif d == 'R':
            directions.remove('L')
        elif d == 'L':
            directions.remove('R')

        if c >= 3:
            directions.remove(d)

        for way in directions:
            if 0 <= dir(way,i,j,c,d)[1] < len(s) and 0 <= dir(way,i,j,c,d)[2] < len(s[0]):
                
                #If we reach the end of the grid
                if dir(way,i,j,c,d)[1] == target[0] and dir(way,i,j,c,d)[2] == target[1]:
                    return(s[dir(way,i,j,c,d)[1]][dir(way,i,j,c,d)[2]] + costs[active[1:]], active)

                new_cost = s[dir(way, i, j, c, d)[1]][dir(way, i, j, c, d)[2]] + costs[active[1:]]
                
                if dir(way, i, j, c, d) not in costs or new_cost < costs[dir(way, i, j, c, d)]:
                    costs[dir(way, i, j, c, d)] = new_cost
                 
                    ol.append((new_cost, way, dir(way, i, j, c, d)[1], dir(way, i, j, c, d)[2], dir(way, i, j, c, d)[3]))

                
```
