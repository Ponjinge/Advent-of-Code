```python
#Boilerplate 1 

s = []

with open('./input.txt', 'r') as file:
    file_content = file.read().strip()  
individual_strings = file_content.split('\n')
for i, individual_str in enumerate(individual_strings, 1):
    word = [*individual_str]
    s.append(word)
```

```python
for i in range(len(s)):
    for j in range(len(s[0])):
        s[i][j]= int(s[i][j])
```

```python
def nv(t):
    return cool[dir_numerate(t[0])][t[1]][t[2]]
```

```python
def dir(way,i,j):
    if way =='U':
        return((way, i-1,j)) 
    if way =='D':
        return((way, i+1,j))
    if way =='L':
        return((way, i,j-1))
    if way =='R':
        return((way, i,j+1))
```

```python
def record_move(way, i, j, new):
    
    if new == False:
        prevmoves[dir_numerate(way)][dir(way,i,j)[1]][dir(way,i,j)[2]].pop(0)
        prevmoves[dir_numerate(way)][dir(way,i,j)[1]][dir(way,i,j)[2]].pop(0)
        prevmoves[dir_numerate(way)][dir(way,i,j)[1]][dir(way,i,j)[2]].pop(0)
        
    prevmoves[dir_numerate(way)][dir(way,i,j)[1]][dir(way,i,j)[2]].append(prevmoves[d][i][j][1])
    prevmoves[dir_numerate(way)][dir(way,i,j)[1]][dir(way,i,j)[2]].append(prevmoves[d][i][j][2])
    prevmoves[dir_numerate(way)][dir(way,i,j)[1]][dir(way,i,j)[2]].append(way)
```

```python
def dir_translate(d):
    if d ==0:
        return 'U'
    elif d ==1:
        return 'D'
    elif d ==2:
        return 'L'
    elif d ==3:
        return 'R'
```

```python
def dir_numerate(way):
    if way =='U':
        return 0 
    if way =='D':
        return 1
    if way =='L':
        return 2
    if way =='R':
        return 3
```

```python
#Initiation

start = ('U',0,0) # Starting position 

ol = [start] # Open List
cl = [] # Closed List 
c=0 # Same direction count 


# Create an empty table for the cost to move to a space 
cool = [[[[]for _ in range(len(s))] for _ in range(len(s))] for _ in range(4)]

# Create an empty table for the 3 previous directions to get to a space
prevmoves = [[[[]for _ in range(len(s))] for _ in range(len(s))] for _ in range(4)]

# We start at 0 with no real previous directions
cool[0][0][0]=0
cool[1][0][0]=0
cool[2][0][0]=0
cool[3][0][0]=0
prevmoves[0][0][0]=['A','B','C']
prevmoves[1][0][0]=['A','B','C']
prevmoves[2][0][0]=['A','B','C']
prevmoves[3][0][0]=['A','B','C']

# While there are elements in the open list 
while ol:
    
    #print(ol)
    
    min = 10**10 
    for i in range (len(ol)):
        
        # We treat the lowest cost element in the list
        if nv(ol[i]) + (250 - (ol[i][1]+ol[i][2])/2) < min:
            min = nv(ol[i])+ (250 - (ol[i][1]+ol[i][2])/2)
            min_index = i
    active = ol[min_index] # It becomes the active space 
    
    # We remove it from the open list and add it to the closed list
    cl.append(active)
    ol.pop(min_index) 
    d = dir_numerate(active[0])
    i = active[1]
    j= active[2]
    
    
    # For each possible direction around that space
    
    directions = ['U','D','L','R']
    
    if prevmoves[d][i][j][2] == prevmoves[d][i][j][1] == prevmoves[d][i][j][0]:
        directions.remove(prevmoves[d][i][j][2])
   
    if prevmoves[d][i][j][2] == 'U':
        directions.remove('D')
    if prevmoves[d][i][j][2] == 'D':
        directions.remove('U')
    if prevmoves[d][i][j][2] == 'R':
        directions.remove('L')
    if prevmoves[d][i][j][2] == 'L':
        directions.remove('R')


    for way in  directions :
        
        #print( i, j , way)
        #print(prevmoves[i][j])
        #print(dir(way, i, j))
      
        
       
        # If this square is already in the Open List for this direction
        if dir(way,i,j) in ol:
            
            
            if cool[dir_numerate(way)][dir(way,i,j)[1]][dir(way,i,j)[2]] >= s[dir(way,i,j)[1]][dir(way,i,j)[2]] + cool[d][i][j]:
                    
                cool[dir_numerate(way)][dir(way,i,j)[1]][dir(way,i,j)[2]] = s[dir(way,i,j)[1]][dir(way,i,j)[2]] + cool[d][i][j]
                  
                #Move recording (works)
                record_move(way, i, j, False)
                    
                    
        # If this square is already in the Closed List for this direction            
        elif dir(way,i,j) in cl:
            
            if cool[dir_numerate(way)][dir(way,i,j)[1]][dir(way,i,j)[2]] >= s[dir(way,i,j)[1]][dir(way,i,j)[2]] + cool[d][i][j]:
                      
                cool[dir_numerate(way)][dir(way,i,j)[1]][dir(way,i,j)[2]] = s[dir(way,i,j)[1]][dir(way,i,j)[2]] + cool[d][i][j]
                    
                # Add to open list
                ol.append(dir(way,i,j))
                    
                # Move recording (works)   
                record_move(way, i, j, False)
                    
                    
        # If this space has not been observed yet from this direction           
        elif 0 <= dir(way,i,j)[1] < len(s) and 0 <= dir(way,i,j)[2] < len(s[0]):
           
            cool[dir_numerate(way)][dir(way,i,j)[1]][dir(way,i,j)[2]] = s[dir(way,i,j)[1]][dir(way,i,j)[2]] + cool[d][i][j]

            # Add to open list
            ol.append(dir(way,i,j))
                
            #Move recording (works)
            record_move(way, i, j, True)
       
                
```

```python
def k_record_move(k, way, i, j):
    
    if prevmoves[k][dir(way,i,j)[1]][dir(way,i,j)[2]] != [] :
        prevmoves[k][dir(way,i,j)[1]][dir(way,i,j)[2]].pop(0)
        prevmoves[k][dir(way,i,j)[1]][dir(way,i,j)[2]].pop(0)
        prevmoves[k][dir(way,i,j)[1]][dir(way,i,j)[2]].pop(0)
        
    prevmoves[k][dir(way,i,j)[1]][dir(way,i,j)[2]].append(prevmoves[d][i][j][1])
    prevmoves[k][dir(way,i,j)[1]][dir(way,i,j)[2]].append(prevmoves[d][i][j][2])
    prevmoves[k][dir(way,i,j)[1]][dir(way,i,j)[2]].append(way)
```
