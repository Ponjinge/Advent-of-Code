```python
import re

s = []

with open('./input.txt', 'r') as file:
    file_content = file.read().strip()  # Read the file content and remove leading/trailing whitespaces

# Split the content based on empty lines
individual_strings = file_content.split('\n')

# Process each individual
for i, individual_str in enumerate(individual_strings, 1):
    # Convert lines to integers and add them to the individuals list
    individual = [line for line in re.split('',individual_str)]
    
    s.append(individual)
   
```

```python
# Get the empty rows and columns 
rM = []
rC = []
for i in range (len(s) - 1, -1, -1):
    e = True
    for j in range(len(s[i])):
        if '#'in s[i][j]:
            e = False
    if e == True :
        rM.append(i)

for i in range (len(s[0]) - 1, -1, -1):
    e = True
    for j in range(len(s)):
        if '#'in s[j][i]:
            e = False
    if e == True :
        rC.append(i)   
        
```

```python
# Get the position of all the Galaxies
r = []
for i in range (len(s)):
    for j in range(len(s[i])):
        if s[i][j]=='#':
            r.append([i,j])
r
```

```python
# Account for dimension expansion by changing the positions based on if they are after a warp 
for i in range (len(r)):
    for y in range(len(rM)):
        if r[i][0]>rM[y]:
            
            r[i][0]= r[i][0]+999999 #Replace by 1 for the first question
            
for i in range (len(r)):
    for y in range(len(rC)):
        if r[i][1]>rC[y]:
            r[i][1]= r[i][1]+999999 #Replace by 1 for the first question
```

```python
# Compute the distance for each pair
dist = []
for i in range(len(r)):
    for j in range( i, len(r)):
        dist.append(abs(r[i][0] - r[j][0]) + abs(r[i][1] - r[j][1]))
```

```python
# Get the sum of distances
sum(dist)
```
