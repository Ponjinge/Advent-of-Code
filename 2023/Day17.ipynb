```python
#Max recursion depth
from functools import lru_cache 

@lru_cache(maxsize = None)
def shortest_route(i,j,d,c):
    if i == len(s)-1 and j == len(s[0])-1:
        return 0
    
    elif 0<=i<=140 and 0<=j<=140 :
        if c !=3:
            if d == 0 and i>0:
                x = shortest_route(i-1,j,d,c+1)+s[i-1][j]
            elif d == 1 and j<140:
                x=shortest_route(i,j+1,d,c+1)+s[i][j+1]
            elif d == 2 and i<140:
                x=shortest_route(i+1,j,d,c+1)+s[i+1][j]
            elif d == 3 and j>0:
                x=shortest_route(i,j-1,d,c+1)+s[i][j-1]
        
        d2 = d-1
        if d2 == 0 and i>0:
            x2 = shortest_route(i-1,j,d,1)+s[i-1][j]
        elif d2 == 1 and j<140:
            x2=shortest_route(i,j+1,d,1)+s[i][j+1]
        elif d2 == 2 and i<140:
            x2=shortest_route(i+1,j,d,1)+s[i+1][j]
        elif d2 == 3 and j>0:
            x2=shortest_route(i,j-1,d,1)+s[i][j-1]
                
        d3 = d+1
        if d3 == 0 and i>0:
            x3 = shortest_route(i-1,j,d,1)+s[i-1][j]
        elif d3 == 1 and j<140:
            x3=shortest_route(i,j+1,d,1)+s[i][j+1]
        elif d3 == 2 and i<140:
            x3=shortest_route(i+1,j,d,1)+s[i+1][j]
        elif d3 == 3 and j>0:
            x3=shortest_route(i,j-1,d,1)+s[i][j-1]
        if c !=3:
            return min(x,x2,x3)
        else :
            return min(x2,x3)
    else: 
        return 1000000000000000000
```

```python
def dir(way,i,j):
    if way =='U':
        return((i-1,j)) 
    if way =='D':
        return((i+1,j))
    if way =='L':
        return((i,j-1))
    if way =='R':
        return((i,j+1))
```
