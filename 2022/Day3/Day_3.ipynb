```python
def get_number_from_letter(char):
    if 'a' <= char <= 'z':
        return ord(char) - ord('a') + 1
    elif 'A' <= char <= 'Z':
        return ord(char) - ord('A') + 27
    else:
        # Handle non-alphabetic characters if needed
        return None

# Example usage:
char_a = 'a'
char_z = 'Z'
char_A = 'A'

print(f"{char_a} corresponds to {get_number_from_letter(char_a)}")
print(f"{char_z} corresponds to {get_number_from_letter(char_z)}")
print(f"{char_A} corresponds to {get_number_from_letter(char_A)}")
```
